# Cleaning Data by Function 

```python
##Input a troop name to find consecutive (at least 6 months) data
##Will return a dataframe with months to keep (Keep = 1) and months to discard (Keep == 0)
##Simply need df.loc[df.Keep != 0]

def consec6(troop):
    months = []
    years = []
    count = 0
    tag = 0 ##keeps track if count hit 5 or greater
    
    for i in census.loc[census.Troop.str.contains(troop)].ObsYear.unique():
        years.extend([i]*len(sig_data.get_group((troop,i)).ObsMonth.unique()))
        months.extend(sig_data.get_group((troop,i)).ObsMonth.unique())
        
    df = pd.DataFrame({'Year':years,'Month':months,'Keep':[0]*len(months)})
    
    df['mon_elap'] = df.Month + 12*(df.Year-list(df.Year)[0])

    for x in range(len(months)-1):
        ##Next unique month consecutive number?
        if df.loc[x+1,'mon_elap'] == df.loc[x,'mon_elap']+1:
            count += 1
            ##End of df and count >= 5? Mark as keep
            if (x == len(months)-2) & (count >= 5):
                df.loc[x-count+1:x+1,'Keep'] = 3
                #print('loc 5')
            #print('loc 1')
    
        ##No longer consecutive and under 6 months? Start over.
        elif (df.loc[x+1,'mon_elap'] != df.loc[x,'mon_elap']+1) & (count < 5):
            count = 0 
            #print('loc 2')

        ##6 months or over and no longer consecutive? Mark as keep
        elif (df.loc[x+1,'mon_elap'] != df.loc[x,'mon_elap']+1) & (count >= 5):
            df.loc[x-count:x,'Keep'] = 1
            count = 0
            #print('loc 3')
            
        ##All Keep entries are 0 at very end?    
        if (np.sum(df.Keep) == 0) & (count >= 5) & (x == len(months)-2):
            df.loc[x-count:x+1,'Keep'] = 2
            #print('loc 4')
            

            
    #print('count is {}'.format(count))

    return df
```

```python
def data_clean(df,op_trp=0):
    ##Function cleans the entire data set such that each troop has at least six consecutive months of observations
    ##Input data set (df) and returns cleaned data set (df)
    
    ##Contains option to clean one troop out of data set (for testing)
    ##Enter op_trp value (string) to clean specific troop
    
    ##Add cleaning column
    df.loc[:,'cleaner'] = 0
    
    ##Create coluimn of tuple that contains (obsMonth,ObsYear)
    df['Tuple'] = list(zip(list(df.ObsMonth),list(df.ObsYear)))
    
    ##Iterate through all troops, clean data, and update cleaner column
    troops = df.Troop.unique()
    
    ##For all troops
    if op_trp == 0:
        for troop in troops:
            
            ##Start cleaning function
            clean = consec6(troop)
            clean = clean.loc[clean.Keep != 0]
            clean['test'] = list(zip(list(clean.Month),list(clean.Year)))
            
            ##Copy of data for cleaner columns (used to update df)
            wd = df.loc[df.Troop.str.contains(troop)]
            
            try:
                ##Find the obs tuples in copied data that are also in cleaned data
                wd = wd[wd['Tuple'].isin(clean.test)]
                
                ##Change intersected data's cleaner column to 1
                wd.loc[:,'cleaner'] = 1
                
                ##Update final df
                df.update(wd)
                
            except:
                   pass
                
    ##Repeat steps for singular troop
    else:
        troop = op_trp
        clean = consec6(troop)
        clean = clean.loc[clean.Keep != 0]
        clean['test'] = list(zip(list(clean.Month),list(clean.Year)))

        wd = df.loc[df.Troop == troop]
        try:
            wd = wd[wd['Tuple'].isin(clean.test)]
            wd.loc[:,'cleaner'] = 1
        
            df.update(wd)
        except:
            pass
        
        
    df = df.loc[df.cleaner == 1]
    
    return df

data_clean(census,'MOW')
```
