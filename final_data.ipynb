```python
import pandas as pd
from matplotlib import pyplot as plt
import datetime
import matplotlib as mpl
import numpy as np
import math
import pylab as pl
import math
```

```python
df = pd.read_csv('../02-24-21/interpreted_data_2.28.21_for gabrielle.csv')
df.loc[:,'Date'] = pd.to_datetime(df.Date)
df.loc[:,'newT'] = round(df.newT)
df.loc[:,'newT'] = df.newT.apply(int)
```

```python
pete = pd.read_csv('C:/Users/GabsS/Desktop/Mongoose Modeling/clean_pete_data.csv')
pete = pete.loc[:,['ObservationID','Troop','T','A','S','J']]

#fixing clerical error
pete.loc[512,['S','J']] = 0
def confirm_age(row):
    
    if row['T'] != row['newT']:
        row['A'] = np.nan
        row['S'] = np.nan
        row['J'] = np.nan
        return row
    
    else:
        try: #take age_class values from matching ObservationIDs
            row['A'] = float(pete.loc[pete.ObservationID == row['ObservationID']]['A'])
            row['S'] = float(pete.loc[pete.ObservationID == row['ObservationID']]['S'])
            row['J'] = float(pete.loc[pete.ObservationID == row['ObservationID']]['J'])
            return row
        
        except: #if there are no matching ObservationIDs leave age_class value empty
            row['A'] = np.nan
            row['S'] = np.nan
            row['J'] = np.nan
            return row
        
```

```python
new_data = df.apply(confirm_age,axis=1)
```

```python
new_data = new_data.loc[:,['Date','Troop','tdate','tdint','quarter','newT','A','S','J']]
new_data = new_data.rename(columns={'newT':'T'})
new_data
```
